```python
import psycopg2
conn = psycopg2.connect(
    host="localhost",
    port=8812,  # Default PostgreSQL wire port in QuestDB
    database="qdb",
    user="admin",
    password="quest"  # Only if auth is enabled
)

df = pd.read_sql("SELECT * FROM binance_depth ORDER BY timestamp DESC LIMIT 50", conn)
print(df)
```

```python
import json
df['bid_data'] = df['bid_json'].dropna().apply(json.loads)
df
```

```python
def calculate_mid_price(bid_str, ask_str):
    try:
        bids = json.loads(bid_str)
        asks = json.loads(ask_str)
        best_bid = float(bids[0][0]) if bids else None
        best_ask = float(asks[0][0]) if asks else None
        if best_bid is not None and best_ask is not None:
            return (best_bid + best_ask) / 2
    except Exception:
        return None

df['mid_price'] = df.apply(lambda row: calculate_mid_price(row['bid_json'], row['ask_json']), axis=1)
df
```

```python
import ta
from ta.momentum import RSIIndicator
from ta.trend import MACD

# Calculamos RSI sobre los mid_price
rsi = RSIIndicator(close=df['mid_price'], window=3)
df['rsi'] = rsi.rsi()

# Calculamos momentum como diferencia simple
df['momentum'] = df['mid_price'].diff()

# Aceleración como la diferencia del momentum (segunda derivada)
df['acceleration'] = df['momentum'].diff()
```
